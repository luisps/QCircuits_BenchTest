```python
import csv
import sys

import matplotlib.pyplot as plt
```

```python
path = "/Users/psantos/Feynman-MC-QSimulator-Data/Feynman-MC-QSimulator-Data"

circuitID = ["41120","41256","41512","411024","412048","414096"]
qubits = ["120","256","512","1024","2048","4096"]

algorithms = "BD_MIS"

istate = "0"

ostate = "1023"

samples = ["27","28","25","24","24","25"]

threads = "48"

files = [path+"/circuit_"+NcircuitID + ".data_stats_" + algorithms + "_" + istate + "_" + ostate + "_" + Nsamples + "_" + threads + ".csv" for NcircuitID,Nsamples in zip(circuitID,samples)]
```

```python
# Read CSVs

data = {}
for ndxfile, file in enumerate(files):
    qbs = qubits[ndxfile] 
    data[qbs] = {}
    with open(file) as csvfile:
        csvreader = csv.reader(csvfile, delimiter=',')
        for rowndx, row in enumerate(csvreader):
            #print (row)
            if rowndx==0: continue     # 1st row is header
            elif rowndx==1:
                true_w = complex(float(row[0]),float(row[1]))
                data[qbs]["true_w"] = true_w            
            elif rowndx==2:     # 3rd row is header
                dataTitles = [s.strip() for s in row]
                # add 3 more attributes
                #dataTitles.append ("variance")
                #dataTitles.append ("L2")
                #dataTitles.append ("efficiency")
                for title in dataTitles:
                    data[qbs][title] = [] 
            else:
                for titlendx, title in enumerate(dataTitles):
                    data[qbs][title].append(float(row[titlendx])) 
    #print (dataTitles)
    
```

```python
## Plot allows specification of dependent and in dependent variables
# choose each from
'n_samples', 'time', 'n_paths', 'n_NZero_paths', 'estimateR', 'estimateI', 'varianceR', 'varianceI'
# independent variable
X_var = 'n_samples'
# dependent variable
Y_var = 'varianceR'

# log X axis ?
log_X_axis = True
# log Y axis ?
log_Y_axis = True
```

```python
fig, ax = plt.subplots()

if log_X_axis:
    # convert x-axis to Logarithmic scale
    plt.xscale("log")
else:
    plt.xscale("linear")


if log_Y_axis:
    # convert y-axis to Logarithmic scale
    plt.yscale("log")
else:
    plt.yscale("linear")

# for each algorithm
for qbs in qubits:
    # Data for plotting
    X = data[qbs][X_var]
    Y = data[qbs][Y_var]

    ax.plot(X, Y, label = qbs)

ax.set(xlabel=X_var, ylabel=Y_var,
       title='Large Hidden Shift Circuits')
ax.grid()
ax.legend ()


plt.show()
```

```python
img = path + "/LargeHSA_" + algorithms + "_" + istate + "_" + ostate + "_" + X_var + "_VS_" + Y_var + ".png"
fig.savefig(img)
```
